5. task 1 : create a code using map() function under lambda condition

```python
input = [1, 2, 3, 4, 5]
inc = 2

list_inc = lambda input, inc: list(map(lambda x: x + inc, input))

list_inc(input, inc)
```

6. task 2 :  create a code using  reduce() function under lambda condition
#from functools import reduce()

```python
from functools import reduce

input = [1, 2, 3, 4, 5]

sum_result = reduce(lambda x, y: x + y, input)
print(sum_result)
```
