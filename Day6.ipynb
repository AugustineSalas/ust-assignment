1. You are given a nested dictionary representing folders 
and file sizes.
Write a recursive function calculate_size(folder) that 
returns the total size of all files.
 
Input:
folder = {
    "docs": {
        "resume.pdf": 120,
        "cover.docx": 80
    },
    "images": {
        "photo1.png": 200,
        "photo2.png": 300
    }
}
 
Output:700
 
Rules:
-Use recursion
-If value is int → file size
-If value is dict → folder
-No global variables

```python
def calculate_size(folder):
    if type(folder) is int:
        return folder
    return sum(map(calculate_size, folder.values()))

folder = {
    "docs": {
        "resume.pdf": 120,
        "cover.docx": 80
    },
    "images": {
        "photo1.png": 200,
        "photo2.png": 300
    }
}

print(calculate_size(folder))
```

2.Write a function audit_change(data) that:
-Appends 100 if input is a list
-Adds key "updated": True if input is a dictionary
-Tries to modify an integer (observe behavior)
 
Input :
x = 10
lst = [1, 2]
d = {"a": 1}
 
Expected Observation:
-x remains unchanged
-lst changes
-d changes
 
Explain why?

```python
def audit_change(data):
    if type(data) is int:
        data += 10
    elif type(data) is list:
        data.append(100)
    elif type(data) is dict:
        data.update({"updated": True})

x = 10
lst = [1, 2]
d = {"a": 1}

audit_change(x) # Expecting no change because integers are immutable
print(x)

audit_change(lst) # Expecting the list to be modified because lists are mutable
print(lst)

audit_change(d) # Expecting the dictionary to be modified because dictionaries are mutable
print(d)
```

4. Write a recursive function validate_config(config) that ensures:
- No value is None
- Nested dictionaries are validated fully
 
Input :
config = {
    "db": {
        "host": "localhost",
        "port": None
    }
}
 
Output: False

```python
def validate_config(config:dict):
    if config is None:
        return False
    if type(config) is dict:
        return all(map(validate_config, config.values()))
    else:
        return True
    
config = {
    "db": {
        "host": "localhost",
        "port": None
    }
}

print(validate_config(config))
```

5. Create a function discount(rate) that returns another function that:
-Accepts price
-Applies discount
 
Example:
ten_percent = discount(10)
ten_percent(200)
 
Output: 180

```python
def discount(rate):
    def apply_discount(price):
        return price * (1 - rate / 100)
    return apply_discount

ten_percent = discount(10)
ten_percent(200)
```

7. Write a recursive function count_leaves(node) that counts the 
leaf nodes in a tree.
 
Input:
tree = {
    "A": {
        "B": {},
        "C": {
            "D": {}
        }
    }
}
 
Output: 2

```python
def count_leaves(node):
    if len(node) == 0:
        return 1
    return sum(map(count_leaves, node.values()))

tree = {
    "A": {
        "B": {},
        "C": {
            "D": {}
        }
    }
}

print(count_leaves(tree))
```
