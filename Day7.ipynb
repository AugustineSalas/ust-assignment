1. Maximum of Three Numbers
Without using max():
-Compare three numbers
-Print the largest
-If all equal → print "All numbers are equal"

```python
def max_num(a, b, c):
    if a == b and b == c:
        return "All numbers are equal"
    return a if a > b and a > c else b if b > c else c

print(max_num(10, 20, 30))
print(max_num(10, 10, 10))
```

2.ATM Transaction
 
Inputs: Balance, Withdrawal amount
Conditions:
- Amount > balance → “Insufficient funds”
- Amount ≤ 0 → “Invalid amount”
- Amount % 100 != 0 → “Enter multiples of 100”
Else → “Transaction successful”

```python
def transaction(balance, withdrawal_amount):
    if withdrawal_amount > balance:
        return "Insufficient funds"
    elif withdrawal_amount <= 0:
        return "Invalid amount"
    elif withdrawal_amount % 100 != 0:
        return "Enter multiples of 100"
    else:
        return "Transaction successful"
    
print(transaction(1000, 500))
print(transaction(1000, 1500))
print(transaction(1000, -100))
print(transaction(1000, 50))
    
```

3.Password Strength Checker
 
Check password:
-Length ≥ 8
-Contains digit
-Contains uppercase
-Contains special character
Print: Strong / Medium / Weak

```python
def password_check(password):
    strength = 0
    if len(password) >= 8:
        strength += 1
    if any(char.isupper() for char in password):
        strength += 1
    if any(char.isdigit() for char in password):
        strength += 1
    if any(char in "!@#$%^&*()-+" for char in password):
        strength += 1

    if strength in (0, 1):
        return "Weak"
    elif strength in (2, 3):
        return "Medium"
    else:
        return "Strong"
    
print(password_check("Password1!"))
print(password_check("pass"))
print(password_check("Password"))
print(password_check("Password1"))
```

4.Student Result System
 
Input: Marks in 5 subjects
Conditions:
- Any subject < 35 → Fail
- Avg ≥ 75 → Distinction
- Avg ≥ 60 → First Class
Else → Pass

```python
def student_result(marks:tuple):
    if any(mark < 35 for mark in marks):
        return "Fail"
    avg = sum(marks) / 5
    if avg >= 75:
        return "Distinction"
    elif avg >= 60:
        return "First Class"
    else:
        return "Pass"
    
print(student_result((80, 70, 90, 60, 75)))
print(student_result((50, 40, 60, 70, 80)))
print(student_result((80, 70, 40, 60, 35)))
print(student_result((30, 70, 90, 60, 75)))
```

**When to use `elif` instead of multiple `if`?**  
- Use `elif` when conditions are mutually exclusive or when you want to check them sequentially (stopping at the first true condition). 
- Multiple `if` statements check all conditions independently, which canlead to unintended multiple executions (enters all true condition).

**Why nested `if` increases complexity?**  
- Nested `if` statements create deeper indentation levels.
- making code harder to read, debug, and maintain. 
- raise the risk of errors (e.g., mismatched braces).

10.Simulate a menu-driven system:
1. Add
2. Subtract
3. Exit
Use an infinite loop and Break only when user selects Exit
 


```python
while True:
    try:
        num1 = int(input("First number: "))
        num2 = int(input("Second number: "))
    except:
        print("Invalid input\n Do you wish to continue? (y/n):")
        if input()[0] == 'y':
            continue
        else:
            print("Program Ended")
            break

    match input("Operations(Add/Subtract/Exit): ")[0]:
        case "A":
            print(num1, '+', num2, '=', num1+num2)
        case "S":
            print(num1, '-', num2, '=', num1-num2)
        case "E":
            print("Program Ended")
            break
        case default:
            print(f"{default} is an invalid operation")
```

11.ATM Session Simulation
Simulate ATM behavior:
-Infinite loop waiting for user
-Limit wrong PIN attempts to 3
-Exit session after success or failure
 


```python
def session_simulation_atm(user_input):
    pin = 1234
    count = 1
    while True:
        if user_input == pin:
            return "Success"
        elif count == 3:
            return "Failed 3 times\nTry again Later"
        else:
            print("Login Failed", 3-count, "times remaining")
            user_input = int(input("Enter pin: "))
            count += 1

print(session_simulation_atm(int(input("Enter pin: "))))
 
```

12.Log Processing - Given server logs:
logs = ["200 OK", "404 ERROR", "500 ERROR", "200 OK", "403 ERROR"]
Count each type of status
Stop processing if "500 ERROR" occurs

```python
def log_processing():
    logs = ["200 OK", "404 ERROR", "500 ERROR", "200 OK", "403 ERROR"]
    status_count = {}
    for i in logs:
        if i == "500 ERROR":
            return status_count
        if i not in status_count:
            status_count[i] = 1
        else:
            status_count[i] += 1

print(log_processing())
```
